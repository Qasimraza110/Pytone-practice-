```python
# group by pandas
```

```python
coffeerating.groupby('owner')
```

```python
coffeerating.groupby('owner').size()
```

```python
coffeerating.groupby('owner').count()
```

```python
# Multiple stats (mean, max, min) for each owner
coffeerating.groupby('owner')['total_cup_points'].agg(['mean', 'max', 'min'])
```

```python
# Number of farms per owner
coffeerating.groupby('owner')['farm_name'].nunique()
```

```python
# Top 10 owners with the most entries
coffeerating.groupby('owner').size().sort_values(ascending=False).head(10)
```

```python
# Show all missing rows grouped by owner
coffeerating[coffeerating.isnull().any(axis=1)].groupby('owner').size()
```

```python

# Get the top scoring farm for each owner
coffeerating.loc[
    coffeerating.groupby('owner')['total_cup_points'].idxmax(),
    ['owner', 'farm_name', 'total_cup_points']
]
```
