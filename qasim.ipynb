```python
print ("practice code ")
a = int(input("Enter first number: "))
b = int(input("Enter second number: "))
op = input("Enter operator (+, -, *, /): ")

if op == "+":
    print("Result:", a + b)
elif op == "-":
    print("Result:", a - b)
elif op == "*":
    print("Result:", a * b)
elif op == "/":
    print("Result:", a / b)
else:
    print("Invalid operator")
```

```python
students = {
    "Ali": [85, 90, 78],
    "Sara": [88, 92, 80],
    "Qasim": [75, 70, 82]
}

for name, marks in students.items():
    avg = sum(marks) / len(marks)
    print(name, "average marks =", round(avg, 2))
```
